```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# x = np.arange(1,6)
# y = np.array([3,4,2,4,5])
ds = pd.read_csv('headbrain.csv')
x = ds['Head Size(cm^3)'].values
y = ds['Brain Weight(grams)'].values
```

```python
mean_x = x.mean()
mean_y = y.mean()
numerator = 0
denominator = 0
for i in range(0,len(ds)):
    out_1 = x[i] - mean_x
    out_2 = y[i] - mean_y
    temp_1 = out_1 * out_2
    numerator += temp_1
    out_3 = out_1 ** 2
    denominator += out_3
m = numerator / denominator

# finding y intercept
c = mean_y - (m * mean_x)

# finding predicted values for y with given x
pred_y = []

for i in range(0,len(ds)):
    temp_2 = m * x[i] + c
    pred_y.append(temp_2)

# finding r square

numrtr = 0
denntr = 0 
for i in range(0,len(ds)):
    temp_3 = (pred_y[i] - mean_y) ** 2
    numrtr += temp_3
    temp_4 = (y[i] - mean_y) ** 2
    denntr += temp_4

r_2 = numrtr / denntr
print(r_2)
```

```python
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
plt.scatter(x,y, marker = '*',label = 'headbrain')
plt.plot(x,pred_y, color = 'blue')
plt.legend()
plt.show()
```

```python
x = x.reshape(-1, 1)
reg = LinearRegression()
reg = reg.fit(x,y)
pred_y = reg.predict(x)

mse = mean_squared_error(y, pred_y)
rmse = np.sqrt(mse)
r2_score = reg.score(x,y)


print(np.sqrt(mse))
print(r2_score)
```
