```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
df = pd.read_csv('area1.csv')
dummies  = pd.get_dummies(df['town'])
dummies
```

```python
merge = pd.concat([df, dummies], axis = 'columns')
merge
```

```python
final = merge.drop(['town','west windsor'], axis = 'columns')
final
```

```python
x = final[['area','monroe township','robbinsville']].values
y = final['price']
reg = LinearRegression()
```

```python
reg.fit(x,y)
reg.predict([[2600,0,1],[2600,0,0]])
reg.score(x,y)
```

```python
i = df.drop('price', axis = 'columns').values
o = df['price'].values
```

```python
preprocessor = make_column_transformer( (OneHotEncoder(),[0]),remainder="passthrough")
ni = preprocessor.fit_transform(i)
varx = ni[:,1:]
varx
```

```python
reg.fit(varx,o)
reg.predict([[1,0,3100]])
reg.score(varx,o)
```
